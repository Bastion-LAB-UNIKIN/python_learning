```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import *
```

```python
y = np.array([1, 2, 2, 3, 5, 2])
y_pred = np.array([5, 2, 2, 5, 7, 1000])
print('MAE  :', mean_absolute_error(y, y_pred))   # erreur absolue moyenne   =========>  linéaire 
print('MSE  :', mean_squared_error(y, y_pred))     # erreur quadratique moyenne ===== > exponentielle ===> grandes erreur
print('RMSE  :', np.sqrt(mean_squared_error(y, y_pred)))   # sqr de MSA
```

```python
#  ------------ Median Absolute error
```

```python
print('Median abs errr : ', median_absolute_error(y, y_pred))
```

```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LinearRegression
```

```python
iris = load_iris()

x = iris.data
y = iris.target

model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)

print(x.shape)
plt.scatter(x[:, 3], y, label='y')
plt.scatter(x[:, 3], y_pred, alpha=0.8, label='y_pred')
plt.legend()
```

```python
#-----------------                TROUVER LA PERFORMANCE DU MODEL               ---------------------------
```

```python
err_hist = np.abs(y - y_pred)
plt.hist(err_hist, bins=50)
plt.show()                                      # on a une distribution exponentielle
```

```python
#------------------------  le coefficient de détérmination    r**2
```

```python
model.score(x,y)
```

```python
#  avec cross validation
```

```python
cross_val_score(LinearRegression(), x, y, cv=3, scoring='neg_mean_absolute_percentage_error')
```
