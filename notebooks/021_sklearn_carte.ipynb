```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
iris = load_iris()

x = iris.data
y = iris.target

print(x.shape)
plt.scatter(x[:, 0], x[:, 1], c=y, alpha=0.8)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold, LeaveOneOut, ShuffleSplit
```

```python
cv = KFold(5)#, random_state=0)     nombre de split
cross_val_score(KNeighborsClassifier(),x,y , cv=cv)
```

```python
 # Nombre d'évaluation  = nombre de split =========>  Leave One Out
```

```python
cv = LeaveOneOut()
cross_val_score(KNeighborsClassifier(),x,y , cv=cv)
```

```python
#----------- TECHNIQUE  02
```

```python
# Mélange ====> teste =====> regroupe ======> mélange ============> retest
```

```python
cv = ShuffleSplit(4, test_size=0.2)
cross_val_score(KNeighborsClassifier(),x,y , cv=cv)
```

```python
#       TECHNIQUE 03  
```

```python
# Crée des splite avec une petite split proportionnel 
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
cv = StratifiedKFold(4)
cross_val_score(KNeighborsClassifier(),x,y , cv=cv)
```

```python
#-------------    GROUPE  kFOLD ---------------------
```

```python
from sklearn.model_selection import GroupKFold
```

```python
cv = GroupKFold(5).get_n_splits(x, y, groups=x[:, 0])
cross_val_score(KNeighborsClassifier(),x,y , cv=cv)
```
