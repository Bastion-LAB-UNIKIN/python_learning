```python
#conditions if , elif , else

def signe(x):
    if (x>0):
        print(x,"positif")
    elif x== 0:
        print(x, "null ")
    else:
        print(x,"négatif ")
```

```python
signe(-1)
```

```python
def e_potentielle(masse,hauteur,limite,g=9.81):
    E = masse * hauteur * g
    return E < limite
```

```python
e_potentielle(masse=50,hauteur=111,limite=7000)
```
