```python
def trier(nombre):
    liste_1 = []
    liste_2 = []

    for i in range(-10,nombre,2):
        if i >= 0:
            liste_1.append(i)
        else:
            liste_2.append(i)

    classeur = {
        "positif" : liste_1,
        "negatif" : liste_2,
    }
    
    return classeur

trier(19)

```
