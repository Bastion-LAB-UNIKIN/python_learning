dictionnaire comprehensin

```python
prenom = ['etienne', 'exauce' , 'jenovie', 'kerene']
age = [24, 25 , 22, 25]
prenom
```

```python
dico={prenom:age for prenom,age in zip(prenom,age)}
# ici on fait la comprehension pour faire un dictionnaire avec les prenoms et les ages
# la boucle for prenom,age in zip(prenom,age) va parcourir les deux listes prenom et age en même temps et va associer chaque prenom avec son age correspondant
# zip(prenom,age) va créer un itérateur qui va retourner des tuples (prenom, age) pour chaque élément des deux listes
dico
```

```python
dico={prenom:age for prenom,age in zip(prenom,age) if age >= 25}
# ici on fait la comprehension pour faire un dictionnaire avec
#  les prenoms et les ages mais seulement pour les ages qui sont supérieurs ou égaux à 25
dico
```
