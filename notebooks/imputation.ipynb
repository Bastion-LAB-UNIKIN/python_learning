```python
import numpy as np # pour les calculs numériques
import matplotlib.pyplot as plt # pour la visualisation
import pandas as pd # pour la manipulation de données
import seaborn as sns # pour la visualisation de données
from sklearn.linear_model import SGDClassifier # pour la classification linéaire
```

```python
#--------------   SIMPLE  IMPUTER 

"""
ce quoi l'imputation ?
l'imputation est une technique qui permet de remplacer les données manquantes par des valeurs estimées à partir des données disponibles.
il existe plusieurs méthodes d'imputation, mais la plus simple est l'imputation par la moyenne, la médiane ou la valeur la plus fréquente (mode)
 pour les données numériques, et la valeur la plus fréquente pour les données catégorielles.
"""
```

```python
from sklearn.impute import SimpleImputer # pour l'imputation simple des données manquantes
```

```python
x = np.array([[10, 3],
             [0, 4],
             [5, 3],
             [np.nan, 3]])
# ce code crée un tableau numpy avec des données numériques, où la dernière ligne contient une valeur manquante (np.nan)
imputer = SimpleImputer(strategy='mean') # on crée un objet imputer avec la stratégie d'imputation par la moyenne
x_imputed = imputer.fit_transform(x) # on applique l'imputation sur le tableau x et on stocke le résultat dans x_imputed
print(x_imputed) # on affiche le tableau imputé, où la valeur manquante a été remplacée par la moyenne des autres valeurs de la colonne (5.0 dans ce cas)   
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
# ici on crée un objet imputer avec la stratégie d'imputation par la moyenne et en spécifiant que les valeurs manquantes sont représentées par np.nan


imputer.fit_transform(x)
# ici on applique l'imputation sur le tableau x et on retourne le résultat, où la valeur manquante a été remplacée par la moyenne des autres valeurs de la colonne (5.0 dans ce cas)
```

```python
x_test = np.array([[12, 5],
                   [40, 2],
                   [5,  5],
                   [np.nan, np.nan]])

imputer.transform(x_test)
# ici on applique l'imputation sur le tableau x_test en utilisant les statistiques calculées à partir du tableau x, et on retourne le résultat, où les valeurs manquantes ont été remplacées par la moyenne des autres valeurs de chaque colonne (5.0 pour la première colonne et 3.0 pour la deuxième colonne dans ce cas)
```

```python
##---------    KKNImputer
# le KNNImputer est une technique d'imputation qui utilise les k plus proches voisins pour estimer les valeurs manquantes.
# il est basé sur l'idée que les données similaires ont des valeurs similaires, et il utilise les distances entre les points
#  de données pour trouver les voisins les plus proches.
```

```python
from sklearn.impute import KNNImputer
```

```python
x = np.array([[10, 100],
             [2, 30],
             [3, 15],
             [np.nan, 3]])
```

```python
imputer = KNNImputer(n_neighbors=1)
imputer.fit_transform(x)
```

```python
from sklearn.impute import MissingIndicator
```

```python
x = np.array([[10, 100],
             [2, 30],
             [3, 15],
             [np.nan, np.nan]])
```

```python
MissingIndicator().fit_transform(x)
```

```python
from sklearn.pipeline import make_union
```

```python
pipeline = make_union(SimpleImputer(strategy='constant', fill_value=-99),
                      MissingIndicator())

pipeline.fit_transform(x)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import SGDClassifier
import seaborn as sns 
from sklearn.model_selection import train_test_split
```

```python
titanic = pd.read_csv('../datasets/titanic.csv')
# on charge le dataset titanic à partir d'un fichier CSV situé dans le dossier ../datasets/

titanic.head()
# on affiche les 5 premières lignes du dataset titanic pour avoir un aperçu des données
```

```python
x = titanic[['pclass', 'age']]
# on sélectionne les colonnes 'pclass' et 'age' du dataset titanic pour les utiliser comme variables explicatives (features) dans notre modèle de classification

y= titanic['survived']


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state = 5)
# on divise les données en un ensemble d'entraînement (x_train, y_train) et 
# un ensemble de test (x_test, y_test) en utilisant la 
# fonction train_test_split de scikit-learn, avec une taille de test de 20% 
# et une graine aléatoire de 5 pour assurer la reproductibilité des résultats
```

```python
model = make_pipeline( KNNImputer(), SGDClassifier())
```

```python
params = {
    'knnimputer__n_neighbors' : [1, 2, 3, 4] 
}
# ici on définit les paramètres à tester pour le KNNImputer, en 
# spécifiant les différentes valeurs de n_neighbors à essayer (1, 2, 3 et 4)
```

```python
grid = GridSearchCV(model, param_grid=params, cv =5)
# ici on crée un objet GridSearchCV pour effectuer une recherche en grille 
# sur les paramètres du modèle, en utilisant une validation croisée à 5 plis (cv=5)
```

```python
grid
# ici on affiche l'objet GridSearchCV pour voir les paramètres et les options de validation croisée qui ont été définis
```

```python
grid.fit(x_train,y_train)
# ici on entraîne le modèle en utilisant la méthode fit de GridSearchCV, qui effectue la recherche en grille sur les paramètres spécifiés et évalue les performances du modèle à l'aide de la validation croisée, en utilisant les données d'entraînement (x_train, y_train)
```

```python
grid.best_params_
# ici on affiche les meilleurs paramètres trouvés par la recherche en grille,
#  qui sont les valeurs de n_neighbors qui ont donné les meilleures performances 
# lors de la validation croisée
```
