dictionnaire comprehensin

```python
prenom = ['etienne', 'exauce' , 'jenovie', 'kerene']
age = [24, 26 , 22, 25]
```

```python
dico={prenom:age for prenom,age in zip(prenom,age)}
dico
```

```python
dico={prenom:age for prenom,age in zip(prenom,age) if age >= 25}
dico
```
