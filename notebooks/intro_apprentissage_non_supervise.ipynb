```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
#-----------------  K- Means   Clustering

# Chaque point du dataset est affecté au cluster du centroid le plus proche

#Le centroid est déplacé au milieu du cluster. là où se situe la moyenne des points
```

```python
from sklearn.cluster import KMeans
```

```python
#Génération de données

x, y = make_blobs(n_samples=100, centers=3,cluster_std=0.4, random_state=10)
plt.scatter(x[:, 0], x[:, 1])
```

```python
model = KMeans(n_clusters=3)
model.fit(x)
model.predict(x)
plt.scatter(x[:, 0], x[:, 1], c=model.predict(x))

plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], c='r')
model.cluster_centers_  #position final de notre centroide
```

```python
#distance entre le point du cluster est le centroide

model.inertia_
```

```python
# definir le nombre de cluster
```

```python
inertia = []
k_range = range(1, 20)
for k in k_range:
    model = KMeans(n_clusters=k).fit(x)
    inertia.append(model.inertia_)


plt.plot()
```
