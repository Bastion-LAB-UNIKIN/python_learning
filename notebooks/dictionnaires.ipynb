```python
inventaire = {
    "bananes" : 5000,
    "pommes" : 2099,
    "poires" : 56898,
    "cerises" : 2892,
}
# les dictionnaires sont des structures de données en Python qui permettent
#  de stocker des paires clé-valeur, où chaque clé est unique et associée à
#  une valeur. Dans cet exemple, le dictionnaire "inventaire" contient des 
# fruits comme clés (bananes, pommes, poires, cerises) et leurs quantités 
# respectives comme valeurs (5000, 2099, 56898, 2892).
inventaire
```

```python
import numpy as np
# numpy est une bibliothèque Python utilisée pour le calcul scientifique et l'analyse de données.
```

```python
parametres = {
    "w1" : np.random.randn(10, 100),
    "b1" : np.random.randn(10, 1),
    "w2" : np.random.randn(10, 10),
    "b2" : np.random.randn(10, 1),
}
# ce dictionnaire "parametres" contient des clés (w1, b1, w2, b2) associées
#  à des valeurs qui sont des tableaux de nombres aléatoires générés par
#  la fonction np.random.randn(). Ces tableaux peuvent être utilisés pour 
# stocker les poids et les biais d'un réseau de neurones, par exemple.
```

```python
def trier(nombre):
    liste_1 = []
    liste_2 = []

    for i in range(-10,nombre,2):
        if i >= 0:
            liste_1.append(i)
        else:
            liste_2.append(i)

    classeur = {
        "positif" : liste_1,
        "negatif" : liste_2,
    }
    
    return classeur

trier(19)

```
