```python
def initialisations(m,n):
    np.random.seed(0)
    E_1 = np.random.randint(0, 10, [m,n+1])
    E_2 = np.ones((m,1))
    E = np.concatenate((E_1, E_2), axis=1)
    return E

print(initialisations(4,3))
# explication : la fonction initialisations génère une matrice de taille m x (n+1) avec des nombres aléatoires entre 0 et 9, puis ajoute une colonne de 1 à la fin de cette matrice.
    
```
