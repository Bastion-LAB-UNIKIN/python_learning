```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2, 10)
y = x**2
print(x)
```

```python
plt.plot(x, y)  #abcisse et les ordonnés
plt.show()
```

```python
plt.scatter(x, y)  #nuage des points
plt.show()
```

```python
plt.plot(x, y, c='r', lw=4, ls='--')   #c= couleur, lw=épaisseur, ls=style de trait,  label=nombre des courbes
plt.show()
```

###FIGURE

```python
plt.figure()     #répresente la feuille de travail ( dimension) figsize=(12, 8)
plt.plot(x, y, label='quadratique')
plt.plot(x, x**3, label='cubique')
plt.title('figure 1')
plt.xlabel('axe x')
plt.ylabel('axe y')
plt.legend()
plt.show()
plt.savefig('figure.png')  #création d une fonction


plt.figure()
plt.plot(x, y)
```

Une GRILLE DE GRAPHIQUES

```python
plt.subplot(2,1,1)
plt.plot(x, y, c='red')
plt.subplot(2,1,2)
plt.plot(x, x**3, c='blue')
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

```python
fig, ax = plt.subplots(2, 1, sharex=True)
ax[0].plot(x, y)
ax[1].plot(x, np.sin(x))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = {f"experience{i}": np.random.randn(100) for i  in range(4)}
```

```python
def graphique(data):
    n = len(data)
    plt.figure(figsize=(12, 8))
    for k,i in zip(data.keys(), range(1, n+1)):
        plt.subplot(n, 1, i)
        plt.plot(data[k])
        plt.title(k)
        
    plt.show()
    
```

```python
graphique(dataset)
```
