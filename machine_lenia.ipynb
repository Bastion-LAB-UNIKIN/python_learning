```python
# video 14

#matplotlib
import matplotlib.pyplot as plt

x = np.linspace(0, 2, 10)
y = x**2
print(x)

plt.plot(x, y)  #trace le graphique

plt.show()  #affiche le graphique pour autre IDE que Jupyter Notebook
plt.scatter(x, y)  #nuage de points

plt.plot(x,y, c = 'red', lw = 5, ls = "--")  # trace le graphique avec des options de couleur, largeur et style de ligne

plt.figure()#figsize=(1,1))  #définit la taille de la figure
plt.title("Titre du graphique")  #ajoute un titre au graphique

plt.plot(x, x**2, c = 'black', label = "quadratique")  #ajoute une légende
plt.plot(x, x**3, c = 'green', label = "cubique") 
plt.xlabel("axe des x")  #ajoute un label à l'axe des x
plt.ylabel("axe des y")  #ajoute un label à l'axe des y
plt.legend()
plt.savefig("figure_1.png")  #sauvegarde le graphique dans un fichier


plt.figure(figsize=(10,5))  # autre figure
plt.subplot(2, 2, 1)  #créé une figure avec 1 ligne et 2 colonnes, et sélectionne la première sous-figure
plt.plot(x, np.sin(2 * np.pi * x), c='blue')
plt.subplot(2, 2, 2)  #sélectionne la deuxième sous-figure
plt.plot(x, np.cos(2 * np.pi * x), c='orange')
plt.subplot(2, 2, 3)  #retourne à la troisième sous-figure
plt.plot(x, np.tan(2 * np.pi * x), c='red')
plt.subplot(2, 2, 4)  #retourne à la quatrième sous-figure
plt.plot(x, np.exp(x), c='green')
#plt.show()
plt.savefig("figure_2.png")  #sauvegarde le graphique dans un fichier figure_2.png

plt.figure()

# P. O. O - Programmation Orientée Objet

fig, ax = plt.subplots(2, 1, sharex=True)  #créé une figure avec 2 sous-figures partageant le même axe x
ax[0].plot(x, y, c='blue')
ax[1].plot(x, -y, c='red')


```
